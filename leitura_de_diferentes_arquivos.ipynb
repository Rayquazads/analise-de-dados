# Usando banco de dados

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, select, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
dados.head()
```

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
empregados = pd.read_sql_query(query, engine)
empregados
```

```python
empregados.to_sql('empregados', engine, index=False)
```

```python
inspector = inspect(engine)
print(inspector.get_table_names())
```

```python
pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente','Grau_escolaridade','Rendimento_anual'])
```

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query, engine)
```

```python
id_cliente_excluir = 5008804

try:
    # Define a query SQL para deletar o cliente
    query = text(f"DELETE FROM clientes WHERE ID_Cliente = {id_cliente_excluir}")

    # Executa a query no banco de dados
    with engine.connect() as connection:
        connection.execute(query)
        connection.commit()  # Confirma a transação para salvar a alteração

    print(f"Cliente com ID {id_cliente_excluir} excluído com sucesso.")

except sqlalchemy.exc.SQLAlchemyError as e:
    print(f"Erro ao excluir o cliente: {e}")
```

```python
pd.read_sql_table('clientes', engine)
```

```python
query = text(f"UPDATE clientes SET Grau_escolaridade='Ensino superior' WHERE ID_Cliente=5008808")
with engine.connect() as connection:
  connection.execute(query)
  connection.commit()  # Confirma a transação para salvar a alteração
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados_novos = pd.read_csv(url)
dados_novos.head()
```

```python
dados_novos.to_sql('clientes_novos', engine, index=False)
```

```python
inspector = inspect(engine)
print(inspector.get_table_names())
```

```python
query = 'UPDATE clientes_novos SET Rendimento_anual	 = "300000" WHERE ID_Cliente = "6840104 "'
with engine.connect() as connection:
  connection.execute(text(query))
  connection.commit()  # Confirma a transação para salvar a alteração
```

```python
query = 'SELECT * FROM clientes_novos WHERE ID_Cliente = "6840104"'
with engine.connect() as connection:
  cliente = pd.read_sql_query(query, connection)
cliente
```

```python
query = 'DELETE FROM clientes_novos WHERE ID_Cliente = "5008809"'
with engine.connect() as connection:
  connection.execute(text(query))
  connection.commit()  # Confirma a transação para salvar a alteração
```

```python
id_cliente_verificar = 5008804

query = text(f"SELECT * FROM clientes WHERE ID_Cliente = {id_cliente_verificar}")

with engine.connect() as connection:
    cliente = pd.read_sql_query(query, connection)

if cliente.empty:
    print(f"O cliente com ID {id_cliente_verificar} foi excluído.")
else:
    print(f"O cliente com ID {id_cliente_verificar} ainda existe no banco de dados.")
```

```python
novo_cliente = pd.DataFrame({
    'ID_Cliente': [6850985],
    'Idade': [33],
    'Grau_escolaridade': ['Doutorado'],
    'Estado_civil': ['Solteiro'],
    'Tamanho_familia': [1],
    'Categoria_de_renda': ['Empregado'],
    'Ocupacao': ['TI'],
    'Anos_empregado': [2],
    'Rendimento_anual': [290000],
    'Tem_carro': [0],
    'Moradia': ['Casa/apartamento próprio']
})

# Adicionando o novo cliente à tabela 'clientes'
novo_cliente.to_sql('clientes', engine, if_exists='append', index=False)

# Consultando a tabela 'clientes' para verificar se o novo cliente foi adicionado
pd.read_sql_table('clientes', engine)
```
