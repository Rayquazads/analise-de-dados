```python
alunos['Pontos_extras'] = alunos['Notas'] * 0.40

alunos
```

```python
alunos['nota_final']  = alunos['Notas'] + alunos['Pontos_extras']
alunos
```

```python
alunos['Aprovado_final'] = alunos['nota_final'].apply(lambda x: 'True' if x >= 6 else 'False')
alunos
```

```python
selecao = (alunos['Aprovado'] == False) & (alunos['Aprovado_final'] == True)
alunos_reprovados_aprovados = alunos[selecao]
alunos_reprovados_aprovados
```
